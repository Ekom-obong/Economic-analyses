```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ngn_data = pd.read_excel('Copy of Nigerian Data.xlsx')
ngn_data.head
```

```python
print(ngn_data.columns)
```

```python

import matplotlib.pyplot as plt

# Set the Date column as the index for time series plotting
ngn_data['Date'] = pd.to_datetime(ngn_data['Date'])
ngn_data.set_index('Date', inplace=True)

# Create line plots for each variable over time
variables = ['M1 money Supply', 'real exchange rate', 'stock market performance', 'inflation rate (%)']
fig, axes = plt.subplots(nrows=len(variables), ncols=1, figsize=(12, 16))

for i, var in enumerate(variables):
    ngn_data[var].plot(ax=axes[i], title=var)
    axes[i].set_xlabel('Date')
    axes[i].set_ylabel(var)
plt.savefig('variable visuals')

plt.tight_layout()
plt.show()
```

```python
# Create scatter plots to visualize relationships between pairs of variables
pairs = [
    ('M1 money Supply', 'inflation rate (%)'),
    ('M1 money Supply', 'real exchange rate'),
    ('stock market performance', 'inflation rate (%)'),
    ('real exchange rate', 'stock market performance')
]

fig, axes = plt.subplots(nrows=len(pairs), ncols=1, figsize=(12, 16))

for i, (x, y) in enumerate(pairs):
    axes[i].scatter(ngn_data[x], ngn_data[y])
    axes[i].set_xlabel(x)
    axes[i].set_ylabel(y)
    axes[i].set_title(f'{x} vs. {y}')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
# Assuming `data` is a DataFrame containing the variables
ngn_data = ngn_data.drop(columns =['Month','Year'])
correlation_matrix = ngn_data.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.savefig('correlation matrix')
plt.show()
```

```python
# Lag the M1 money supply by 6 months
ngn_data['M1_lagged'] = ngn_data['M1 money Supply'].shift(6)
lagged_correlation = ngn_data[['M1_lagged', 'inflation rate (%)']].corr(method='pearson')
print(lagged_correlation)
sns.heatmap(lagged_correlation, annot=True, cmap='coolwarm')
plt.title('lagged_correlaition')
plt.show()
```

```python
# Lag the stock market performance by 6 months
ngn_data['Stock_lagged'] = ngn_data['stock market performance'].shift(6)
lagged_correlation = ngn_data[['Stock_lagged', 'inflation rate (%)']].corr(method='pearson')
print(lagged_correlation)
```

```python
# Lag the real exchange rate by 6 months
ngn_data['exchange_lagged'] = ngn_data['real exchange rate'].shift(6)
lagged_correlation = ngn_data[['exchange_lagged', 'inflation rate (%)']].corr(method='pearson')
print(lagged_correlation)
```
